```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import os
import glob
import pandas as pd
```

```python
pmf_data_1 = np.loadtxt('./pmf_139-147.txt', usecols=None)
pmf_data_1 = pmf_data_1[:150,:]
print (pmf_data_1)
#Check data using view graphs
plt.figure(figsize=(10,8))
plt.plot(pmf_data_1[:,0],pmf_data_1[:,1], '-', linewidth = 5, label = 'test')
plt.xlabel('Q_global', fontsize = 20)
plt.ylabel('Free energy (kcal/mol)', fontsize = 20)
plt.legend(prop={'size': 20})
plt.show()
```

```python
pmf_data_2 = np.loadtxt('./pmf_4BEX_assembly.txt', usecols=None)
pmf_data_2 = pmf_data_2[101:,:]
print (pmf_data_2)
for i in range(len(pmf_data_2)):
    pmf_data_2[i,1]=pmf_data_2[i,1]-10.6
print (pmf_data_2)
#Check data using view graphs
plt.figure(figsize=(10,8))
plt.plot(pmf_data_2[:,0],pmf_data_2[:,1], '-', linewidth = 5, label = 'test')
plt.xlabel('Q_global', fontsize = 20)
plt.ylabel('Free energy (kcal/mol)', fontsize = 20)
plt.legend(prop={'size': 20})
plt.show()
```
