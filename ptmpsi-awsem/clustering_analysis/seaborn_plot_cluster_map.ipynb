```python
import scipy 
from scipy.cluster import hierarchy
from scipy.spatial import distance
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
q_data_path = './139-147_rep1.dat'
```

```python
_1D_Q_data = np.loadtxt(q_data_path)
_1D_Q_data_squareform = scipy.spatial.distance.squareform(_1D_Q_data)
qw = _1D_Q_data_squareform
q = _1D_Q_data
```
