```python
# This notebook is to concatenate the trajectories to be used

import os
import mdtraj as md

# Specify the path to the main folder containing the subfolders
main_folder = "/home/chengxuan/Dropbox/Cofilin_oligomer/umbrella_sampling/screening_rerun/139-147/"
output_folder = "./"
# Create an empty list to store the trajectories


# Iterate over the subfolders
for i in range(1, 6):
    subfolder_path = os.path.join(main_folder, f"rep{i}")
    trajectories = []
    # Iterate over the subsubfolders
    for j in range(24, 40):
        subsubfolder_path = os.path.join(subfolder_path, str(j))
        pdb_file_path = os.path.join(subsubfolder_path, "step1.pdb")

        # Load the PDB trajectory using MDTraj
        trajectory = md.load(pdb_file_path)

        # Use stride to subsample the frames
        subsampled_trajectory = trajectory[::10]

        # Append the subsampled trajectory to the list
        trajectories.append(subsampled_trajectory)
        print (subsubfolder_path + " added")
    
    concatenated_trajectory = md.join(trajectories)
        
    output_file_path = os.path.join(output_folder, f"rep{i}/139-147_rep{i}.pdb")
    concatenated_trajectory.save(output_file_path)
        
    print (subfolder_path + " completed")
```
